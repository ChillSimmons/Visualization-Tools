# Seaborn & Matplotlib
### Visualizations can be helpful in conveying insights from analysis
* matplotlib - 2D plotting library
* seaborn - high-level interface for statistical data

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Create a Violinplot - with matplotlib

```python
# initialize object for figure, axes

fig, ax = plt.subplots()

tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

ax.violinplot(tips['total_bill'], vert=False)
plt.show()
```

### Create a Violinplot - with seaborn
* also uses plt.show( )

```python
import seaborn as sns

tips = sns.load_dataset('tips')

sns.violinplot(x = 'total_bill', data = tips)
plt.show()
```

### Seaborn has built-in datasets - load using load_dataset( )
* however, works best with pandas DataFrames, and Arrays

### Construct a Swarmplot

```python
iris = sns.load_dataset('iris')

sns.swarmplot(x = 'species', y = 'petal_length', data = iris)
plt.show()
```

### Construct a Factorplot
* a categorical plot - here, a bar plot

```python
titanic = sns.load_dataset('titanic')

g = sns.factorplot('class', 'survived', 'sex', data=titanic, kind='bar',
                  palette='muted', legend=False)
plt.show()
```

### Matplotlib can utilize Seaborn's colors as a Colormap
* to define a color - use color_palette( )
* for number of colors - use n_colors argument (here, use 5)

```python
import numpy as np
from matplotlib.colors import ListedColormap

# define a list length, N - where n_colors number of integers vary
# construct colormap

N = 500

current_palette = sns.color_palette('muted', n_colors = 5)
cmap = ListedColormap(sns.color_palette(current_palette).as_hex())

# initialize 2 sets of data - assume there are 5 labels

data1 = np.random.randn(N)
data2 = np.random.randn(N)

colors = np.random.randint(0,5,N)
```

### Construct a Scatterplot

```python
plt.scatter(data1, data2, c=colors, cmap=cmap)

plt.colorbar()
plt.show()
```

### Seaborn plots can be Scaled
* various contexts available - for talks, notebook, poster, and papers
* create larger or smaller plots/elements - use set_context( )

```python
# reset default params - setup plot to manipulate

sns.set()

# Paper Context

sns.set_context('paper')

iris = sns.load_dataset('iris')

sns.swarmplot(x = 'species', y = 'petal_length', data=iris)
plt.show()
```

* scale plot elements - using font_scale, & rc

```python
sns.set_context('paper', font_scale=3, rc={'font.size':8,
                                           'axes.labelsize':5})

sns.swarmplot(x='species', y='petal_length', data=iris)
plt.show()
```

* set( ) can be used alone to make changes - instead of set_context

```python
sns.set(rc={'font.size':8, 'axes.labelsize':12})

sns.swarmplot(x='species', y='petal_length', data=iris)
plt.show()
```

### Seaborn visualizations have 2 types of functions - axes-level, figure-level
* Axes-level include - regplot( ), boxplot( ), kdeplot( ), etc.
    * these take an ax argument - return an Axes object
    * set figure size - using plt.subplots( ) & figsize

* Figure-level include - lmplot( ), factorplot( ), jointplot( ), etc.
    * these have total control over the figure and need to be set up
    * set figure size - using size & aspect


### Temporarily set a plot style - for figures with axes of different styles
* using axes_style( ) in a with statement
* use this with plt.subplot( )

```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')

# set first style

with sns.axes_style('white'):
    plt.subplot(211)
    sns.swarmplot(x='species', y='petal_length', data=iris)

# set the second style

plt.subplot(212)
sns.violinplot(x = 'total_bill', data=tips)

plt.show()
```

### Rotate label text in Seaborn - Construct a lmplot
* requires working on Figure-level function

```python
x = 10 ** np.arange(1, 10)
y = x * 2

data = pd.DataFrame(data={'x': x, 'y': y})

grid = sns.lmplot('x', 'y', data, size=7, truncate=True,
                  scatter_kws={'s': 100})

# rotate x-axis labels

grid.set_xticklabels(rotation=30)

plt.show()
```

### Set xlim or ylim in Seaborn - Construct a Boxplot
* this can be done at the Axes-level

```python
tips = sns.load_dataset('tips')

ax = sns.boxplot(x='total_bill', data=tips)

ax.set(xlim=(0, 100))

plt.show()
```

* to be done at the Figure-level - use axes & ax

```python
x = 10 ** np.arange(1, 10)
y = x * 2

data = pd.DataFrame(data={'x': x, 'y': y})

lm = sns.lmplot('x', 'y', data, size=7, truncate=True,
                  scatter_kws={'s':100})

axes = lm.ax

axes.set_ylim(-1000000000,)

plt.show()
```

* FacetGrid - uses fig, to expose underlying figure

```python
lm = sns.lmplot('x', 'y', data, size=7, truncate=True,
                  scatter_kws={'s':100})

fig = lm.fig

fig.suptitle('This is the Figure Title', fontsize=12)

plt.show()
```

### Set a Log Scale - Construct a Regplot( )
* use regplot( ) - for simple regression
* use Axes - to set scale

```python
x = 10 ** np.arange(1, 10)
y = x * 2

data = pd.DataFrame(data={'x': x, 'y': y})

fig, ax = plt.subplots()

ax.set(xscale='log', yscale='log')

sns.regplot('x', 'y', data, ax=ax, scatter_kws={'s':100})

plt.show()
```

* for Figure-level functions - use set( ) method of the FacetGrid object

```python
titanic = sns.load_dataset('titanic')

g = sns.factorplot('class', 'survived', 'sex', data=titanic, kind='bar',
                  size=6, palette='muted', legend=True)

g.set(yscale='log')

plt.show()
```

### A Title can also be added using set_title( )
* for Axes-level functions

```python
tips = sns.load_dataset('tips')

ax = sns.boxplot(x='total_bill', data=tips)

ax.set_title('This is a Boxplot')

plt.show()
```

* Figure-level functions can use fig - as well as work via Axes

```python
g = sns.FacetGrid(tips, col='sex', row='smoker', margin_titles=True)

g.map(plt.scatter, 'total_bill', 'tip')

g.fig.suptitle('Here is the Title')

plt.show()
```
